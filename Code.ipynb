### Removing non-english apps

According to the American Standard Code for Information Interchange, ord (latin characters) range between 0 and 127. We can use that to detect if the name of an app is in English or in another language. Lets say that a name with more than 3 special characters is not english. By doing so we avoir getting rid of apps name 'Instachat üòú' or 

```python
def english_name_detector(word) : 
    count_special_characters = 0
    for character in word :
        if ord(character) > 127 :
            count_special_characters += 1 
            if count_special_characters > 3 :
                return False  
        
    return True
```

## <span style='color:darkblue'> FREQUENCY TABLES </span>

We want to find app that are profitable in both apple and google store. 
We can select many criteria to say if an app will be profitable, for example : Number of user, number or ratings, Hight rating, etc

```python
# apple , genre = index 11 ,
# google, genre = index 9, category = index 1
```

```python
def freq_table(dataset, index):
    
    dictionnary = {}
    
    for row in dataset : 
        key = row[index]
        if key in dictionnary :
            dictionnary[key] += 1
        else : 
            dictionnary[key] = 1
            
    for key in dictionnary :
        dictionnary[key] = (dictionnary[key]/len(dataset))*100
     
    return dictionnary
```

```python
freq_table(android_free, 9)
```

```python
def display_table(dataset, index):
    table = freq_table(dataset, index)
    table_display = []
    for key in table:
        key_val_as_tuple = (table[key], key)
        table_display.append(key_val_as_tuple)

    table_sorted = sorted(table_display, reverse = True)
    for entry in table_sorted:
        print(entry[1], ':', entry[0])
```

```python
freq_genre_apple = freq_table(apple_free, 11)
```

```python
for genre in freq_genre_apple:
    total = 0
    len_genre = 0
    
    for app in apple_free :
        genre_app = app[11]
        if genre_app == genre :
            user_ratings = float(app[5])
            total += user_ratings
            len_genre +=1
    avg_ratings = total/len_genre
    print(genre,':', avg_ratings )
```
